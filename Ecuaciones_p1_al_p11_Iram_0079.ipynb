Problema 1: $x + 8 = 15$
* Planteamiento: Un número aumentado en 8 nos da 15.
* Ecuación: $x + 8 = 15$
* Despeje: El 8 que suma pasa restando.$$x = 15 - 8$$Solución: $x = 7$

```python
#Problema 1
# Ecuación: x + 8 = 15
resultado = 15
termino_independiente = 8

# Despejamos x restando el 8 al resultado
x = resultado - termino_independiente

print(f"Resultado Problema 1: x = {x}")
```

Problema 3: $3x + 4x = 21$
* Planteamiento: El triple de un número más cuatro veces ese mismo número es 21.* Ecuación: $3x + 4x = 21$
* Despeje: Primero sumamos las $x$: $3 + 4 = 7$.$$7x = 21$$
* Luego, el 7 que multiplica pasa dividiendo:$$x = \frac{21}{7}$$Solución: $x = 3$

```python
#Problema 3
# Ecuación: 3x + 4x = 21 -> (3+4)x = 21
suma_coeficientes = 3 + 4
total = 21

# Despejamos x dividiendo el total entre la suma de los coeficientes
x = total / suma_coeficientes

print(f"Resultado Problema 3: x = {int(x)}")
```

Problema 5: $6a - 2a + a = 20$
* Planteamiento: Seis veces un número, menos dos veces ese número, más una vez el mismo número es igual a 20.
* Ecuación: $6a - 2a + a = 20$
* Despeje:
* Agrupamos las "a": $6 - 2 + 1 = 5$.$$5a = 20$$
* El 5 pasa dividiendo:$$a = \frac{20}{5}$$Solución: $a = 4$

```python
#Problema 5
# Ecuación: 6a - 2a + 1a = 20
# Agrupamos los coeficientes de 'a'
coeficientes_a = 6 - 2 + 1
total = 20

# Despejamos 'a' dividiendo el total entre el coeficiente resultante
a = total / coeficientes_a

print(f"Resultado Problema 5: a = {int(a)}")
```

Problema 7: $4x + 3 = 2x + 11$
* Planteamiento: Tenemos incógnitas en ambos lados. Hay que mover las $x$ a la izquierda y los números a la derecha.
* Ecuación: $4x + 3 = 2x + 11$
* Despeje: 1.  
* El $2x$ pasa restando a la izquierda: $4x - 2x + 3 = 11$   
* El $3$ pasa restando a la derecha: $4x - 2x = 11 - 3$
* Resolvemos ambos lados: $2x = 8$
* Dividimos: $x = \frac{8}{2}$.
* Solución: $x = 4$

```python
# Problema 7
# Ecuación: 4x + 3 = 2x + 11
# Movemos términos: 4x - 2x = 11 - 3
diferencia_x = 4 - 2
diferencia_numeros = 11 - 3

# Despejamos x
x = diferencia_numeros / diferencia_x

print(f"Resultado Problema 7: x = {int(x)}")
```

Problema 9: $9m + 4 = 5m + 20$
* Planteamiento: Similar al anterior, organizamos los términos.
* Ecuación: $9m + 4 = 5m + 20$
* Despeje:
* Pasamos $5m$ restando: $9m - 5m = 20 - 4$
* Resolvemos: $4m = 16$
*  Dividimos: $m = \frac{16}{4}$.
* Solución: $m = 4$

```python
# Problema 9
# Ecuación: 9m + 4 = 5m + 20
# Agrupamos m a la izquierda y números a la derecha: (9-5)m = 20 - 4
coeficiente_final_m = 9 - 5
resultado_numerico = 20 - 4

# Resolvemos para m
m = resultado_numerico / coeficiente_final_m

print(f"Resultado Problema 9: m = {int(m)}")
```

Problema 11: $6a + 5 = 2a + 21$
* Planteamiento: Agrupamos variables y constantes.
* Ecuación: $6a + 5 = 2a + 21$
* Despeje:$6a - 2a = 21 - 5$$4a = 16$$a = \frac{16}{4}$
* Solución: $a = 4$

```python
#Problema 11
# Ecuación: 6a + 5 = 2a + 21
# Agrupamos: (6-2)a = 21 - 5
coeficiente_a = 6 - 2
constante_derecha = 21 - 5

# Calculamos el valor de a
a = constante_derecha / coeficiente_a

print(f"Resultado Problema 11: a = {int(a)}")
```
